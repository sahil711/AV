```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
pd.options.display.max_columns=100
pd.options.display.max_rows=100
```

```python
train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
```

```python
target=train.amount_spent_per_room_night_scaled
```

```python
print train.shape, test.shape
```

```python
train.booking_date=pd.to_datetime(train.booking_date,format='%d/%m/%y')
train.checkin_date=pd.to_datetime(train.checkin_date,format='%d/%m/%y')
train.checkout_date=pd.to_datetime(train.checkout_date,format='%d/%m/%y')
```

```python
test.booking_date=pd.to_datetime(test.booking_date,format='%d/%m/%y')
test.checkin_date=pd.to_datetime(test.checkin_date,format='%d/%m/%y')
test.checkout_date=pd.to_datetime(test.checkout_date,format='%d/%m/%y')
```

```python
print 'min_dates: ', train.booking_date.min(),train.checkin_date.min(),train.checkout_date.min()
print 'max_dates: ', train.booking_date.max(),train.checkin_date.max(),train.checkout_date.max()
```

```python
print 'min_dates: ', test.booking_date.min(),test.checkin_date.min(),test.checkout_date.min()
print 'max_dates: ', test.booking_date.max(),test.checkin_date.max(),test.checkout_date.max()
```

### there are about 12 cases where the checkin date year is 2012 and for those 12 cases checkin_date<booking_date

### DATA IS SPLIT BY THE MEMBER ID, THERE IS NO OVERLAP IN MEMBER ID IN THE TRAIN AND TEST

```python
df=pd.concat((train.iloc[:,:-1],test),axis=0)
```

### base variables

```python
df['checkin_out_delta']= (df.checkout_date - df.checkin_date).dt.days
df['booking_checkin_delta']= (df.checkin_date - df.booking_date).dt.days
df['booking_checkout_delta']= (df.checkout_date - df.booking_date).dt.days

df['booking_mon']= df.booking_date.dt.month
df['booking_day']= df.booking_date.dt.day
df['booking_year']= df.booking_date.dt.year
df['booking_dow']= df.booking_date.dt.dayofweek
df['booking_week_no']= df.booking_date.dt.week
df['booking_day_of_year']= df.booking_date.dt.dayofyear
df['booking_quarter']= df.booking_date.dt.quarter
df['booking_week_of_yr']= df.booking_date.dt.weekofyear

df['checkin_mon']= df.checkin_date.dt.month
df['checkin_day']= df.checkin_date.dt.day
df['checkin_year']= df.checkin_date.dt.year
df['checkin_dow']= df.checkin_date.dt.dayofweek
df['checkin_week_no']= df.checkin_date.dt.week
df['checkin_day_of_year']= df.checkin_date.dt.dayofyear
df['checkin_quarter']= df.checkin_date.dt.quarter
df['checkin_week_of_yr']= df.checkin_date.dt.weekofyear

```

```python
train=df[df.memberid.isin(train.memberid)]
test=df[df.memberid.isin(test.memberid)]

print train.shape,test.shape
```

```python
train['amount_spent_per_room_night_scaled']=target
```
